```python
import pandas as pd
import matplotlib.pyplot as plt

# Declaração do DataFrame / DataFrame Declaration
df = pd.read_csv('dados_municipais_comparativos.csv')
display(df)
```

```python
# Média e a soma da populaão das cidades / Mean and sum of the population of cities

print(f"A média da população das cidade é: {df['População'].mean()}")
print(f"A soma de todos os habitantes de cada cidade é: {df['População'].sum()}")
print(f"Cidade com a maior população: Rio Casca: {df['População'].max()}")
print(f"Cidade com a menor população: Sem-Peixe: {df['População'].min()}") 
```

```python
# Imprime o saldo de cada município / Prints the balance for each municipality

df['Saldo'] = df['Receitas (R$)'] - df['Despesas (R$)']
display(df)
```

```python
# Imprime se houve superávit ou não / Prints whether there was a surplus or not

def situacaoFinanceira(saldo):
    if saldo > 0:
        return "SUPERÁVIT"
    else:
        return "DÉFICIT"


df['Situação'] = df['Saldo'].apply(situacaoFinanceira)
display(df)
```
